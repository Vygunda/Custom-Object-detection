```python
img = cv2.imread('./street_image.jpg')

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('./street_image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')  # Hide axes for a cleaner look
plt.show()
```

```python
# predictions
img_pred = yolo.predictions(img)
```

```python
cv2.imshow('prediction image',img_pred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# alternate approach to restore 
import cv2
from matplotlib import pyplot as plt

# Assuming img_pred is already loaded or generated
img_pred_rgb = cv2.cvtColor(img_pred, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

plt.imshow(img_pred_rgb)
plt.title('Prediction Image')
plt.axis('off')  # Optional: hides the axis for a cleaner image
plt.show()
```

## Real Time Object Detection

```python
cap = cv2.VideoCapture('video.mp4')


while True:
    ret, frame = cap.read()
    if ret == False:
        print('unable to read video')
        break
        
    pred_image = yolo.predictions(frame)
    
    cv2.imshow('YOLO',pred_image)
    if cv2.waitKey(1) == 27:
        break
        
cv2.destroyAllWindows()
cap.release()
```
